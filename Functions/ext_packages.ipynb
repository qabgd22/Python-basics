### MAP function

```python
def currency_formatter(number):
    return '$' + str(number)
```

```python
price_list = []

for i in range(5):
    price_list.append(round(uniform(1, 100),2))
    
price_list

# uniform() function to generate a random floating point number between 1 and 100
```

```python
list(map(currency_formatter, price_list))
```
